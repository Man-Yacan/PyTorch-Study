```python
def get_value(n):
    x = (1+1j)**n
    return x.real, x.imag
```

```python
import matplotlib.pyplot as plt
```

```python
point_coor = [get_value(i) for i in range(1, 11)]
x, y  = zip(*point_coor)

plt.plot(x, y)
plt.show()
```
