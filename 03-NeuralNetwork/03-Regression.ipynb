```python
import torch
import re
import numpy as np
```

```python
# 读取文件
f = open('./data/HousePrice.data').readlines()
data = []
for row in f:
    row_handled = re.sub(r'\s{2,}', ' ', row).strip()  # 列与列之间的空格数不等，全部替换为2个
    data.append(row_handled.split(' '))

data = np.array(data).astype(np.float32)
data
```

```python
# 划分数据集
split_num = 496
X_train = data[:496, :-1]
y_train = data[:496, -1]
X_test = data[496:, :-1]
y_test = data[496:, -1]
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# 定义回归网络
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, 100)
        self.predict = torch.nn.Linear(100, n_output)

    def forward(self, X):
        out_hidden = self.hidden(X)
        res = torch.relu(out_hidden)
        return self.predict(res)

net_er = Net(13, 1)

# loss
loss_fun = torch.nn.MSELoss()

# Optimizer
optimizer = torch.optim.Adam(net_er.parameters(), lr=0.0001)

# training
for i in range(1000):
    X = torch.tensor(X_train, dtype=torch.float32)
    y = torch.tensor(y_train, dtype=torch.float32)
    y_pred = torch.squeeze(net_er.forward(X))
    loss  = loss_fun(y_pred, y) * 0.0000001
    optimizer.zero_grad() # 将优化器的梯度设置为0
    loss.backward()
    optimizer.step()

    print(f'iter: {i}, loss: {loss}')
```

```python
# Saving model
torch.save(net_er, './model/model.pkl')
# loading model
# torch.load('./model/model.pkl')
```
